```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def extract_description(driver):
    """Extracts the full product description by clicking 'Read More'."""
    try:
        # Step 1: Find and click 'Read More' button
        read_more = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CLASS_NAME, "read-more"))
        )
        driver.execute_script("arguments[0].click();", read_more)

        # Step 2: Wait for expanded description to load
        WebDriverWait(driver, 5).until(
            EC.visibility_of_element_located((By.CLASS_NAME, "maxi-desc"))
        )

    except Exception as e:
        print(f"‚ö†Ô∏è Could not expand description: {e}")

    # Step 3: Extract full description
    try:
        description_element = driver.find_element(By.CLASS_NAME, "maxi-desc")
        description = description_element.text.strip()
    except:
        description = "No description available"

    return description

# Example Usage
driver = webdriver.Chrome()  # Make sure to install chromedriver first
driver.get("https://example.com/product-page")  # Replace with the actual product URL

full_description = extract_description(driver)
print(full_description)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def accept_cookies(driver):
    """Finds and clicks the 'Accept Cookies' button if present."""
    try:
        cookie_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Accept')]"))
        )
        cookie_button.click()
        print("‚úÖ Cookies accepted")
    except Exception:
        print("‚ö†Ô∏è No cookie popup found or couldn't accept cookies.")

def extract_description(driver):
    """Extracts the product description, handling both cases (with and without 'Read More')."""
    try:
        # Step 1: Check if "Read More" button exists
        read_more_buttons = driver.find_elements(By.CLASS_NAME, "read-more")

        if read_more_buttons:
            print("üîπ 'Read More' found, clicking...")
            driver.execute_script("arguments[0].click();", read_more_buttons[0])

            # Wait for expanded description
            WebDriverWait(driver, 5).until(
                EC.visibility_of_element_located((By.CLASS_NAME, "maxi-desc"))
            )
        else:
            print("üîπ No 'Read More' button found, extracting visible description.")

        # Step 2: Extract description (from either expanded or default text)
        try:
            description_element = driver.find_element(By.CLASS_NAME, "maxi-desc")
        except:
            description_element = driver.find_element(By.CLASS_NAME, "teaser")  # Default short description
        
        return description_element.text.strip()
    
    except Exception as e:
        print(f"‚ö†Ô∏è Error extracting description: {e}")
        return "No description available"

# Initialize WebDriver
driver = webdriver.Chrome()  # Make sure chromedriver is installed
driver.get("https://www.bebitalia.com/en-us/en-charles-carre-divani.html")  # Replace with actual product URL

accept_cookies(driver)  # Handles cookies before interacting
full_description = extract_description(driver)
print("üìù Full Description:\n", full_description)

driver.quit()
```
